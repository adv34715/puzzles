### Merge the data
<p>I read the description of the database, I found there are three main tables: accused, case and trail. So I just merge all the three tables using python Pandas.</p>

<p>There are 205 columns in total. And I delete 5 columns which is 
<ul>

<li>ExecutionDate</li>
<li>FledDate</li>
<li>TrialDate</li>
<li>WatchingDate.</li>
</ul>
Values of these columns are strings. I delete this 5 columns because each of these columns has another column that has the same values but the values are date instead of string. 
</p>

### Output the data
<p>The data is output into a csv file named 'witchData.csv'.</p>

```python
import pandas as pd
```

```python
case = pd.read_csv('WDB_Case.csv')
accused = pd.read_csv('WDB_Accused.csv')
trail = pd.read_csv('WDB_Trial.csv')
```

```python
data_caseAcc = pd.merge(case,accused,how='left',on = 'AccusedRef')
```

```python
data = pd.merge(trail,data_caseAcc,how='left',on = 'CaseRef')
```

```python
data.head()
```

```python
col_data = list(data.columns.values)
col_data.sort()
#col_data
```

```python
def dropDate():
    for c in col_data:
        if 'Date' in c:
            if (c+'_as_date') in col_data:
                print(c, 'is dropped')
                data.drop(c,axis = 1, inplace = True)
```

```python
dropDate()
```

```python
data = data[:100]
```
